```python
import plotly.express as px
import pandas as pd

# Derreter o DataFrame para ter uma linha por m√™s por munic√≠pio por ano
rmvp_melted = rmvp.melt(
    id_vars=['Munic√≠pio de notifica√ß√£o', 'Codigo Munic√≠pio', 'Ano'],
    value_vars=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    var_name='M√™s',
    value_name='Casos'
)

# Mapear nomes dos meses para n√∫meros para facilitar a ordena√ß√£o temporal
meses_map = {'Jan': 1, 'Fev': 2, 'Mar': 3, 'Abr': 4, 'Mai': 5, 'Jun': 6, 'Jul': 7, 'Ago': 8, 'Set': 9, 'Out': 10, 'Nov': 11, 'Dez': 12}
rmvp_melted['M√™s_num'] = rmvp_melted['M√™s'].map(meses_map)

# Criar uma coluna de data combinando Ano e M√™s
# Usamos o 28 como dia para evitar problemas com meses que n√£o t√™m 31 dias
rmvp_melted['Data'] = pd.to_datetime(rmvp_melted['Ano'].astype(str) + '-' + rmvp_melted['M√™s_num'].astype(str) + '-28')

# Converter a coluna 'Casos' para num√©rica (pode haver strings como '-')
rmvp_melted['Casos'] = pd.to_numeric(rmvp_melted['Casos'], errors='coerce').fillna(0).astype(int)

# Ordenar por Data para o gr√°fico de linhas
rmvp_melted = rmvp_melted.sort_values(by='Data')

# Criar o gr√°fico de linhas interativo com Plotly Express
fig = px.line(rmvp_melted, x='Data', y='Casos', color='Munic√≠pio de notifica√ß√£o',
              title='Evolu√ß√£o Mensal dos Casos de Dengue por Munic√≠pio no Vale do Para√≠ba (2007-2025)')

# Atualizar o layout para garantir que o eixo X mostre o intervalo de datas completo
fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

# Exibir o gr√°fico
fig.show()

# Opcional: verificar o intervalo de datas no dataframe
print("Intervalo de datas no dataframe:")
print(rmvp_melted['Data'].min())
print(rmvp_melted['Data'].max())
```
