```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```python
url_pop = 'https://gist.githubusercontent.com/OliveiraGmo/a86b8d4d2c3eb294fc8da27fe0756dc5/raw/4c8edc00871e26b3cfa693b4532b8c2c4ac36e91/dados_populacao.csv'
url= 'https://gist.githubusercontent.com/OliveiraGmo/ea2ba75a8f323d8eacbc468587ba1d08/raw/48ee8e733cc2af4ccf10a96cbb8601a2aa5ff11f/rmvp.csv'
```

```python
populacao = pd.read_csv(url_pop,sep=',')
dados = pd.read_csv(url,sep=',')
```

```python
populacao.head()
```

```python
populacao.columns
```

```python
populacao.drop(columns=['2025'], inplace=True)
```

```python
dados.head()
```

* Alterando os nomes das colunas para que fiquem iguais para facilitar a análise.

```python
populacao.rename(columns={'Municipio':'Município de notificação','Cód.Município':'Codigo Município'}, inplace=True)
```

```python
populacao.columns
```

```python
order = ['Município de notificação','Codigo Município', '2006', '2007', '2008',
       '2009', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022', '2023', '2024']
populacao = populacao[order]
```

```python
populacao.head(5)
```

```python
populacao['Município de notificação'] = populacao['Município de notificação'].str.upper()
```

```python
populacao.drop(columns=['2006'], inplace=True)
```

```python
populacao['2010'] = (populacao['2011'] + populacao['2009'])//2
```

```python
dados = dados.sort_values(by='Ano')
```

```python
dados.head(5)
```

```python
dados['Populacao'] = 0

for index, row in dados.iterrows():
    codigo_municipio = row['Codigo Município']
    ano = str(row['Ano'])
    if ano in populacao.columns:
        populacao_municipio = populacao[populacao['Codigo Município'] == codigo_municipio]
        if not populacao_municipio.empty:
            dados.loc[index, 'Populacao'] = populacao_municipio[ano].iloc[0]

display(dados.head())
```

```python
dados.head(5)
```

```python
for meses in ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']:
    dados[meses] = (dados[meses] / dados['Populacao'])*100000
    dados[meses] = dados[meses].replace([np.inf, -np.inf], np.nan)
    dados[meses] = dados[meses].fillna(0)
    dados[meses] = dados[meses].astype(int)

```

```python
# Filter out rows where 'Ano' is 2025
dados = dados[dados['Ano'] != 2025]
```

```python
month_columns = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
dados['Total'] = dados[month_columns].sum(axis=1)
dados.reset_index(drop=True, inplace=True)
```

```python
dados.reset_index(drop=True, inplace=True)
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.line(dados, x='Ano', y='Total',color='Município de notificação',
              title='Incidência Anual de Dengue no RMVPLN de 2007 à 2024')
fig.update_layout(xaxis_title='Ano', yaxis_title='Incidência')


fig.add_trace(go.Scatter(x=[dados['Ano'].min(), dados['Ano'].max()], y=[100, 100], mode='lines',
                         line=dict(color="green", width=2, dash="dash"), name='Média Incidência', showlegend=True))
fig.add_trace(go.Scatter(x=[dados['Ano'].min(), dados['Ano'].max()], y=[300, 300], mode='lines',
                         line=dict(color="orange", width=2, dash="dash"), name='Alta Incidência', showlegend=True))
fig.add_trace(go.Scatter(x=[dados['Ano'].min(), dados['Ano'].max()], y=[1000, 1000], mode='lines',
                         line=dict(color="red", width=2, dash="dash"), name='Altíssima Incidência', showlegend=True))


fig.update_layout(showlegend=True)

fig.show()
```

```python
dados_melted = dados.melt(
    id_vars=['Município de notificação', 'Codigo Município', 'Ano'],
    value_vars=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    var_name='Mês',
    value_name='Incidencia'
)

meses_map = {'Jan': 1, 'Fev': 2, 'Mar': 3, 'Abr': 4, 'Mai': 5, 'Jun': 6, 'Jul': 7, 'Ago': 8, 'Set': 9, 'Out': 10, 'Nov': 11, 'Dez': 12}
dados_melted['Mês_num'] = dados_melted['Mês'].map(meses_map)

dados_melted['Data'] = pd.to_datetime(dados_melted['Ano'].astype(str) + '-' + dados_melted['Mês_num'].astype(str) + '-28')

dados_melted['Incidencia'] = pd.to_numeric(dados_melted['Incidencia'], errors='coerce').fillna(0).astype(int)

dados_melted = dados_melted.sort_values(by='Data')

fig = px.line(dados_melted, x='Data', y='Incidencia', color='Município de notificação',
              title='Evolução Mensal da Incidência de Dengue por Município no RMVPLN (2007-2024)')


fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=False #Aqui caso queira da para adicionar um barra  slider para exibir os dados
        ),
        type="date"
    )
)


fig.add_trace(go.Scatter(x=[dados_melted['Data'].min(), dados_melted['Data'].max()], y=[100, 100], mode='lines',
                         line=dict(color="green", width=2, dash="dash"), name='Média Incidência', showlegend=True))
fig.add_trace(go.Scatter(x=[dados_melted['Data'].min(), dados_melted['Data'].max()], y=[300, 300], mode='lines',
                         line=dict(color="orange", width=2, dash="dash"), name='Alta Incidência', showlegend=True))
fig.add_trace(go.Scatter(x=[dados_melted['Data'].min(), dados_melted['Data'].max()], y=[1000, 1000], mode='lines',
                         line=dict(color="red", width=2, dash="dash"), name='Altíssima Incidência', showlegend=True))


fig.show()
```
